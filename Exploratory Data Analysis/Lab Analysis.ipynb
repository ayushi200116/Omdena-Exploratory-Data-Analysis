```python
#renaming
labanalysis.rename(columns={'%Time': 'Time'}, inplace=True)
labanalysis.head()
```

```python
#changing time format
labanalysis['Time'] = pd.to_datetime(labanalysis['Time'], unit='D', origin='1899-12-30')
```

```python
#correlation
plt.figure(figsize=(40,30))
sns.heatmap(labanalysis.corr(), annot=True)
```

```python
correlations = labanalysis.corr()
cols = correlations
for i in cols:
    for j in cols:
         if i!=j and abs(correlations.loc[i, j])>= 0.6:
            print(f"{i} and {j} correlated with value {correlations.loc[i, j]}")
```

```python
#scatter plot with respect to time 
columns=['irr_PH', 'irr_EC', 'irr_NH4', 'irr_K', 'irr_Na', 'irr_Ca','irr_Mg', 'irr_Si', 'irr_NO3', 'irr_Cl', 'irr_SO4', 'irr_HCO3','irr_PO4', 'irr_Fe', 'irr_Mn', 'irr_Zn', 'irr_B', 'irr_Cu', 'irr_Mo','drain_PH', 'drain_EC', 'drain_NH4', 'drain_K', 'drain_Na', 'drain_Ca','drain_Mg', 'drain_Si', 'drain_NO3', 'drain_Cl', 'drain_SO4','drain_HCO3', 'drain_PO4', 'drain_Fe', 'drain_Mn', 'drain_Zn','drain_B', 'drain_Cu', 'drain_Mo']
for i in columns:
    labanalysis.plot(kind='scatter',x='Time', y=i, alpha=0.5, c='Time',cmap=plt.get_cmap("jet"))
```
